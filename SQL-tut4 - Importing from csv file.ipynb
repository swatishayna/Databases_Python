# Opening and Reading the file
- FunPart - dont forget to check what row[0] , row[1],... hold.
- check row[4][2:], etc

```python
cur = mydb.cursor()
with open("carbon_nanotubes.csv", "r") as data:
    data_csv = csv.reader(data, delimiter = ";")
    next(data_csv)
    l_left = []
    l_inserted = []
    query = "INSERT INTO Carbon_Nanotubes VALUES(%s,%s,%s,%s,%s,%s,%s,%s)"
    for count,j in enumerate(data_csv):
        try:
            cur.execute(query , j)
            mydb.commit()
            l_inserted.append(count)
            print("l_inserted",count)
            #print(j)
        except:
            l_left.append(count)
```
